```python
from pathlib import Path 

images_dir = Path("./rural_and_urban_photos")
val_dir_rural= images_dir / "val/rural"
val_dir_ubran=  images_dir / "val/urban" 
```

```python
val_rural_images = [i for i in val_dir_rural.glob("*.jpeg") ]+[f for f in  val_dir_rural.glob("*.jpg")]
val_ubran_images = [i for i in val_dir_ubran.glob("*.jpeg") ]+[f for f in val_dir_ubran.glob("*.jpg")]
```

```python
for i,file in enumerate(val_ubran_images):
    image =Image.open(file)
    plt.subplot(3,3,i+1)
    plt.imshow(image)
    plt.axis(False)
    plt.title(f"Image {i}\n{image.width}X{image.height}",fontsize=9)
# This is the key line:
plt.tight_layout()    
plt.show()
    
```

```python
import random # Import the random module

def plot_images(imgs_paths): 
    num_to_plot = min(9, len(imgs_paths)) # Determine how many images to plot (max 9 or fewer if not enough)

    # Randomly sample 'num_to_plot' unique image paths
    samples = random.sample(imgs_paths, k=num_to_plot)

    samples
    for i,file in enumerate( samples):
        if i >= 9 : 
            break
        image =Image.open(file)
        plt.subplot(3,3,i+1)
        plt.imshow(image)
        plt.axis(False)
        plt.title(f"Image {i}\n{image.width}X{image.height}",fontsize=9)
    # This is the key line:
    plt.tight_layout()    
    plt.show()
```

```python
train_dir_rural = images_dir / "train/rural"
train_dir_urban=  images_dir / "train/urban"
train_rural_images = [i for i in train_dir_rural.glob("*.jpeg") ]+[f for f in train_dir_rural.glob("*.jpg")]
train_urban_images = [i for i in train_dir_urban.glob("*.jpeg") ]+[f for f in train_dir_urban.glob("*.jpg")]
```

```python
plot_images(train_rural_images)
```
