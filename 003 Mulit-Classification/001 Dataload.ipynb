```python
import keras
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
assert x_train.shape == (50000, 32, 32, 3)
assert x_test.shape == (10000, 32, 32, 3)
assert y_train.shape == (50000, 1)
assert y_test.shape == (10000, 1)
```

```python
 #These are the standard class names for CIFAR-10, ordered by their integer label (0-9)
class_names = [
    'airplane',
    'automobile',
    'bird',
    'cat',
    'deer',
    'dog',
    'frog',
    'horse',
    'ship',
    'truck'
]
```

```python

print(f"CIFAR-10 Class Names: {class_names}")
print(f"Number of classes: {len(class_names)}")

# Example of a true label
print(f"Label for the first training image (integer): {y_train[0][0]}")
print(f"Corresponding class name: {class_names[y_train[0][0]]}")
```

```python
import matplotlib.pyplot as plt


def show_sample(imgs,labels): 
    plt.figure(figsize=(10,10))
    for i in range(12) :
        plt.subplot(3,4,i+1)
        plt.imshow(imgs[i])
        label_index = labels[i][0]
        plt.title(f"{class_names[label_index]}")
        plt.axis(False)
    plt.tight_layout()
    plt.show()
show_sample(x_train,y_train)   
```
