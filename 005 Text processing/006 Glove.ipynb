```python
def load_glove_embeddings(file_path): 

    embeddings = {} 

    with open(file_path,'r',encoding='utf-8') as f : 
        for line in f: 
            values = line.split()

            word = values[0]

            vector = [float(val) for val in values[1:] ] 
            embeddings[word] = vector 
    return embeddings        
```

```python
file_name = "./glove.6B.100d.txt/glove.6B.100d.txt"
```

```python
glove = load_glove_embeddings(file_name)
```

```python
embedding_dim = 50 


def  text_to_embeddings(text,embeddings=glove): 

    words = text.lower().split()

    embedding_list = [embeddings.get(word,[0]*embedding_dim) for word in words ]

    return embedding_list
```

```python
text_to_embeddings("This is simple text")
```

```python
man = text_to_embeddings("man") 
woman =  text_to_embeddings("woman") 
                         
```

```python
from sklearn.metrics.pairwise import cosine_similarity 

cosine_similarity(man,woman)
```

```python
cosine_similarity(text_to_embeddings("mustafa"),text_to_embeddings("omer")) 
```
